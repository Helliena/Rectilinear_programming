**Задача 1.**

**Постройте схему размера 20 для функции SUM<sub>7</sub>. Напишите для неё программу на Python, подтверждающую корректность схемы.** 

1. Дано 7 бит => максимальная возможная сумма = 7 

   7<sub>10</sub> = 111 <sub>2</sub> => нужно получить 3 цифры двоичной записи

   Заметим, что i-тая цифра в двоичной записи числа = 1 тогда и только тогда, когда имеется нечётное количество слагаемых 2<sup>i</sup>. Таким образом, последняя цифра числа = 1, если $\displaystyle\sum\limits_{i=1}^7 x_{i}$ mod 2 = 1,

   Вторая цифра двоичной записи числа = 1, если $\displaystyle\sum\limits_{1  <= i < j <= 7}^7 ( x_{i} * x_{j})$ mod 2 = 1, 

   Третья цифра двоичной записи числа = 1, если $\displaystyle\sum\limits_{1  <= i < j < k < l <= 7}^7 (x_{i} * x_{j} * x_{k} * x_{l})$ mod 2 = 1. 

   Аналогично мы можем посчитать любую цифру двоичной записи числа суммы n битов. Цифра прод номером k = $\displaystyle\sum\limits_{1  <= i < j < ...  < s < < t <= n}^n x_{i}$ mod 2 = 1, так что в произведении содержится ровно 2^k бит. 


2.  По свойству дистрибутивности: x<sub>1</sub> ${\wedge}$ (x<sub>2</sub> ${\oplus}$ x<sub>3</sub>) = x<sub>1</sub> ${\wedge}$ x<sub>2</sub>  ${\oplus}$   x<sub>1</sub> ${\wedge}$ x<sub>3</sub>. 
Данное свойство верно и для большего числа переменных, что позволяет эффективно рассчитать все 3 цифры двоичной записи суммы 7 бит. 








```
from itertools import product
for x1, x2, x3, x4, x5, x6, x7 in product((0, 1), repeat = 7) :
  x8 = x7 ^ x6
  x9 = x8 ^ x5
  x10 = x9 ^ x4
  x11 = x10 ^ x3
  x12 = x11 ^ x2
  x13 = x12 ^ x1 #последняя цифра
  
  x14 = x1 & x12
  x15 = x2 & x11
  x16 = x3 & x10
  x17 = x4 & x9
  x18 = x5 & x8
  x19 = x6 & x7
  x20 = x19 ^ x18
  x21 = x17 ^ x16
  x22 = x20 ^ x21
  x23 = x14 ^ x15
  x24 = x23 ^ x22 #предпоследняя цифра 

  x25 = x23 & x22
  x26 = x21 & x20
  x27 = x25 ^ x26 #первая цифра

  assert (x1 + x2 + x3 + x4 + x5 + x6 + x7) == x27 * 4 + x24 * 2 + x13
```

1. Вычисление суммы 5 бит за 11 действий вместо 12 позволяет уменьшить размер схемы x<sub>1</sub> + x<sub>2</sub> + x<sub>3</sub> + x<sub>4</sub> + x<sub>5</sub> + x<sub>6</sub> + x<sub>7</sub> до 19 действий:

  (u<sub>1</sub>u<sub>0</sub>)<sub>2</sub> = x<sub>5</sub> + x<sub>6</sub> + x<sub>7</sub>
  
  (v<sub>2</sub>v<sub>1</sub>z<sub>0</sub>)<sub>2</sub> = (x<sub>1</sub> + x<sub>2</sub> + x<sub>3</sub> + <sub>4</sub> + u<sub>0</sub>)

  (w<sub>2</sub>z<sub>1</sub>)<sub>2</sub> = u<sub>1</sub> + v<sub>1</sub>

  z<sub>2</sub> = x<sub>2</sub> ${\oplus}$ w<sub>2</sub>

```
from itertools import product
for x1, x2, x3, x4, x5, x6, x7 in product((0, 1), repeat = 7) :
  x8 = x6 ^ x7 
  x9 = x5 ^ x8 # = u0
  x10 = x6 & x7
  x11 = x5 & x8
  x12 = x10 ^ x11 # = u1
  
  x13 = x1 ^ x2
  x14 = x1 ^ x3
  x15 = x4 ^ x9
  x16 = x3 ^ x13
  x17 = x13 | x14
  x18 = x4 ^ x16
  x19 = x16 ^ x17 
  x20 = x9 ^ x18 #последняя цифра
  x21 = (~x15) & x18
  x22 = x17 ^ x21 # = v1
  x23 = x19 & (~x22) # = v2

  x24 = x12 ^ x22 #предпоследняя цифра
  x25 = x12 & x22 # = w2

  x26 = x23 ^ x25 #первая цифра
  assert(x1 + x2 + x3 + x4 + x5 + x6 + x7 == 4 * x26 + 2 * x24 + x20)
```

```
from itertools import product
for x1, x2, x3, x4, x5, x6 in product((0, 1), repeat = 6) :
  x7 = x1 ^ x2
  x8 = x3 ^ x4
  x9 = x1 ^ x5
  x10 = x3 ^ x5
  x11 = x2 ^ x6
  x12 = x8 ^ x9
  x13 = x8 | x10
  x14 = x7 ^ x13
  x15 = (~x12) & x13
  x16 = (~x11) & x14
  x17 = x11 ^ x15
  x18 = not(x16) and not(x17)
  if (x18) :
    sum = x1 + x2 + x3 + x4 + x5 + x6
    assert(sum % 3 == 0)
```

**Задача 6.**

**Докажите, что для любого n найдётся схема размера не более 2 * n, вычисляющая ${MOD^3_{n}}$**.

2. Докажем, что если мы можем посчитать ${MOD^3_{n}}$ для всех чисел i от 1 до n за количество действий <= 3i, то мы сможем посчитать ${MOD^3_{n + 1}}$ не более, чем за 3 * (n + 1) действий 

3. База индукции

   **n = 0**
   
   t<sub>1</sub> = 0 - **z<sub>1</sub>**

   t<sub>2</sub> = 0 - **z<sub>0</sub>**

   t<sub>3</sub> = ${\lnot}$ t<sub>1</sub> ${\wedge}$ = 1 - всегда => 0 действий

   ---------------------------------------------------------------------

   **n = 1**

   t<sub>1</sub> = x<sub>1</sub> - **z<sub>1</sub>**

   t<sub>2</sub> = 0  - **z<sub>0</sub>**

  ---------------------------------------------------------------------
  
   **n = 2**

   t<sub>1</sub> = x<sub>1</sub> ${\oplus}$ x<sub>2</sub> - **z<sub>1</sub>**

   t<sub>2</sub> = x<sub>1</sub> ${\oplus}$ x<sub>2</sub> - **z<sub>0</sub>**

   t<sub>3</sub> = ${\lnot}$ t<sub>1</sub> ${\wedge}$  - 2 действия

4. Заметим, что можно получить ответ для ${MOD^3_{n}}$, используя ответ для ${MOD^3_{n - 3}}$ всего за 9 промежуточных действий (n + 1 - подсчёт ответа), однако, так как нас не интересует сам ответ для ${MOD^3_{n - 3}}$, важно только z<sub>0</sub>, z<sub>1</sub> того ответа, то мы можем не вычислять ответ для ${MOD^3_{n - 3}}$, а вычислять **только z<sub>0</sub>, z<sub>1</sub>**, а оставшиеся действия использовать для получения ответа ${MOD^3_{n}}$.

  **z<sub>0</sub>, z<sub>1</sub>** - ответы для  ${MOD^3_{n - 3}}$

  **x<sub>1</sub>, x<sub>2</sub>, x<sub>3</sub>** - новые биты

  t<sub>1</sub> = x<sub>1</sub> ${\equiv}$ z<sub>1</sub>

  t<sub>2</sub> = t<sub>1</sub> ${\lor}$ z<sub>0</sub>

  t<sub>3</sub> = t<sub>2</sub> ${\oplus}$ x<sub>2</sub>

  t<sub>4</sub> = t<sub>3</sub> ${\wedge}$ z<sub>0</sub>

  t<sub>5</sub> = x<sub>1</sub> ${\equiv}$ t<sub>4</sub>

  t<sub>6</sub> = ${\lnot}$ t<sub>5</sub> ${\wedge}$ t<sub>2</sub>

  t<sub>7</sub> = t<sub>6</sub> ${\equiv}$ x<sub>3</sub> - **z'<sub>1</sub>**

  t<sub>8</sub> = t<sub>7</sub> ${\equiv}$ t<sub>3</sub>

  t<sub>9</sub> = ${\lnot}$ t<sub>6</sub> ${\wedge}$ t<sub>8</sub> - **z'<sub>0</sub>**

5. Для 

   n = 0 -> 0 действий, 1 строка - ответ

   n = 1 -> 2 действия, 1 строка  - ответ

   n = 2 -> 2 действия, 1 строка  - ответ,

   Для n = k:

   1. k mod 3 = 0 - ответ за 

      k = 3t

      **0 + 9 * t = 9t = 3n действий**

   2. k mod 3 = 1 - ответ за

      k = 3t + 1
      
      **2 + 9 * t + 1 = 9t + 3 = 3n действий**

   3. k mod 3 = 2 - ответ за

      **2 + 9t + 1 = 3n - 3 < 3n действий** 

6. Пример для **n = 15**







   




```
from itertools import product

def f (x1, x2, x3, z0, z1) :
  g1 = (x1 == z1)
  g2 = g1 or z0
  g3 = g2 ^ x2
  g4 = g3 ^ z0
  g5 = (x1 == g4)
  g6 = not(g5) and g2
  g7 = (g6 == x3) #z1 (1)
  g8 = (g7 == g3)
  g9 = not(g6) and g8 #z0 (1)
  return g7, g9

def g (x1, x2, x3) :
  g1 = not(x1) ^ x2
  g2 = (x1 == g1)
  g3 = not(g2) and not(x1)
  g4 = (g3 == x3) #z1 (1)
  g5 = (g4 == g1)
  g6 = not(g3) and g5 #z0 (1)
  return g4, g6

for  x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13, x14, x15 in product((0, 1), repeat = 15) :
  z11, z01 = g(x1, x2, x3)
  z12, z02 = f(x4, x5, x6, z01, z11)
  z13, z03 = f(x7, x8, x9, z02, z12)
  z14, z04 = f(x10, x11, x12, z03, z13)
  z15, z05 = f(x13, x14, x15, z04, z14)
  ans = not(z15) and not(z05)
  sum = x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 + x10 + x11 + x12 + x13 + x14 + x15
  if (sum % 3) :
    mod = 0
  else :
    mod = 1
  assert (mod == ans)
```
